```python
murder_2011 = df_murders[df_murders['YEAR'] == 2011]
murder_2011 = murder_2011.dropna(subset = ['Population'])
state = murder_2011.groupby('STATE')['murder/population'].mean()
state = state.sort_values(ascending=False)
```

```python
max_value = state.max()

# Create bar chart
fig, ax = plt.subplots()
ax.bar(state.index, state)

# Set x-axis labels to rotate for better readability
plt.xticks(rotation=90)

# Set y-axis limits to 0 and the maximum value in the data
ax.set_ylim([0, max_value])
plt.ylabel('Murders/Population')
plt.title('Percentage of murders by population of each state in 2011')

plt.show()
```

```python

import plotly.graph_objs as go

murder_2011 = murder_2011.sort_values(by='murder/population', ascending=False)

# Calculate murder rates per capita
murder_per_capita = murder_2011['murder/population']

# Create bar chart
fig = go.Figure([go.Bar(x=murder_2011['STATE'], y=murder_per_capita)])

# Set labels and title
fig.update_layout(title='Murder/population of all states in 2011',
                  xaxis_title='States in india',
                  yaxis_title='Murders/population')

# Set y-axis limits to 0 and the maximum value in the data
fig.update_yaxes(range=[0, max(murder_per_capita)])

# Show the plot
fig.show()
```
